Import libraries

```python
import pandas as pd
from ks import ks
```

Read made-up data

```python
df = pd.read_csv('ks_example.csv', dtype={'target':'int'})
df.head()
```

# Case 1

The relationship between score and target is negative (*the greater the score, the lower the lower expected probability of event*)

```python
t = ks(score=df['negative_relationship_score'], target=df['target'], bins=20, negative_relationship=True,
       round_intervals=False, print_table=False, write_table=False)
t
```
